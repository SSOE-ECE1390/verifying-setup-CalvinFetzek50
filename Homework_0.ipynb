### Load an (Grayscale) image ('Data/Apollo_11_Launch.jpg')

```python
# import packages
import cv2
import numpy as np
import matplotlib.pyplot as plt

# read image in gradyscale mode
image = cv2.imread('./Data/Apollo_11_launch.jpg',cv2.IMREAD_GRAYSCALE)
```

### View Image

```python
# convert BGR to RGB color format
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image in a window
plt.imshow(image)
```

### Crop Image

```python
# crop the image
cropped_image = image[100:600, 500:700]
plt.imshow(cropped_image)
```

### Edit Brightness of the Image

```python
# edit brightness of the image
# alpha controls contrast (<1: less, >1: more), beta controls brightness (-127<beta<127)
bright_image = cv2.convertScaleAbs(image, alpha=1.2, beta=100)
plt.imshow(bright_image)
```

### Rotate Image

```python
# check image dimensions
print(image.shape)
height, width = image.shape[:2]
center = (width / 2, height / 2)

# rotate image by 45 degrees with scale factor of 1.0
rotate_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image, rotate_matrix, (width, height))

# Display the rotated image
plt.imshow(rotated_image)
```

### Save Modified Image

```python
# save the modified image
cv2.imwrite('./Data/Modified_Apollo_11_launch.jpg', cv2.cvtColor(rotated_image, cv2.COLOR_RGB2BGR))
```
